```python
#Now we ca begin to extract vaccine slots for these dates in the date_list
dist_id = 434
min_age=55
```

```python
##Vaccine availability by District
for input_dates in date_str_list:
    URL="https://cdn-api.co-vin.in/api/v2/appointment/sessions/public/calendarByDistrict?district_id={}&date={}".format(dist_id,input_dates)
    data = requests.get(URL,headers=headers)
    data_json = data.json()
    if data_json["centers"]:
        print("Centers Available on: {}".format(input_dates))
        for center in data_json["centers"]:
            for session in center["sessions"]:
                if session['min_age_limit']<=min_age:
                    print("\t","Center Name:",center['name'])
                    print("\t","District Name:",center['district_name'])
                    print("\t","Block Name:",center['block_name'])
                    print("\t","Price:",center['fee_type'])
                    print("\t","Vaccine:",session['vaccine'])
                    print("\t","Availability: ",session['available_capacity'])
                    print("\n\n")  
                    
    else:
        print("No available slots on : {}".format(input_dates))
```

```python
##District IDs 
URL4 = "https://cdn-api.co-vin.in/api/v2/admin/location/districts/16"
requests.get(URL4,headers=headers).json() ##these districts are in state id 16
```

```python
state_num = []
district_id=[]
district_name = []
```

```python
for i in State_id:
    URL = "https://cdn-api.co-vin.in/api/v2/admin/location/districts/{}".format(i)
    data = requests.get(URL,headers=headers)
    data_json = data.json()
    for x in range(len(data_json['districts'])):
        district_id.append(data_json['districts'][x]['district_id'])
        district_name.append(data_json['districts'][x]['district_name'])
    state_num.append(len(data_json['districts']))
            
```

```python
district_df= pd.DataFrame(data = [district_id,district_name]).transpose()
district_df.columns = ['District_ID','District_Name']
```

```python
final_district_mapping_df = pd.concat([state_df,district_df],axis=1).sort_values(by= ['State_ID','District_Name']).reset_index().drop('index',axis=1)
final_district_mapping_df = final_district_mapping_df[['State_ID','District_ID','District_Name','State_Name']]
final_district_mapping_df.head(10)
```

```python
import json
import numpy as np
import pandas as pd
import plotly.express as px
import plotly.io as pio
```

```python
india_states = json.load(open("states_india.geojson", "r"))
#Source = https://un-mapped.carto.com/tables/states_india/public/map
```

```python
state_id_map = {}
for feature in india_states["features"]:
    feature["id"] = feature["properties"]["state_code"]
    state_id_map[feature["properties"]["st_nm"]] = feature["id"]
```

```python
state_id_map
```

```python
def change_keyname(df, oldkey, newkey):
    df[newkey] = df.pop(oldkey)

change_keyname(state_id_map,"Andaman & Nicobar Island","A & N Islands")
change_keyname(state_id_map,"Dadara & Nagar Havelli","Dadra & Nagar Haveli")
change_keyname(state_id_map,"NCT of Delhi","Delhi")
change_keyname(state_id_map,"Arunanchal Pradesh","Arunachal Pradesh")
```

```python
vaccine_final["id"] = vaccine_final["State/UT"].apply(lambda x: state_id_map[x])
vaccine_final.head()
```

```python
fig = px.choropleth(
    vaccine_final,
    locations="id",
    geojson=india_states,
    color="Total Doses",
    hover_name="State/UT",
    hover_data=["Total Doses"],
    title="India State-wise Vaccination",
)
fig.update_geos(fitbounds="locations", visible=True)
fig.show()
```
